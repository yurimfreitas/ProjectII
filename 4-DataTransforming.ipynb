# 1 Import from pkl files

```python
import joblib
import matplotlib.pyplot as plt

df_sales = joblib.load('df_sales_silver.pkl')
df_product = joblib.load('df_product_silver.pkl')
df_cities = joblib.load('df_cities_silver.pkl')
df_forecast_revenue = joblib.load('df_forecast_revenue_silver.pkl')
```

# 2 Joining data

```python
import pandas as pd

# Join the dataframes on the 'store_id' column (sales and cities)
df_sales_comp = pd.merge(df_sales, df_cities, on='store_id')
```

```python
df_sales_comp
```

```python
# Join the dataframes on the 'product_id' column (sales_comp and products)
df_sales_comp = pd.merge(df_sales_comp, df_product, on='product_id')
```

```python
df_sales_comp
```

## 3 Plot after merge the data to keep understanding some distribution

```python
# Group by city_code and store_id, then sum the sales
grouped_data = df_sales_comp.groupby(['city_code', 'store_id'])['sales'].sum().reset_index()

# Find the top 5 cities by total sales
top_cities = grouped_data.groupby('city_code')['sales'].sum().nlargest(5).index

# Filter the data to include only the top 5 cities
top_cities_data = grouped_data[grouped_data['city_code'].isin(top_cities)]

# Find the top 5 stores by sales within each city
top_stores_per_city = (
    top_cities_data.groupby('city_code')
    .apply(lambda x: x.nlargest(5, 'sales'))
    .reset_index(drop=True)
)

# Pivot the data to get a matrix with cities as rows and stores as columns
pivot_data = top_stores_per_city.pivot(index='city_code', columns='store_id', values='sales').fillna(0)

# Calculate the total sales for each store
store_totals = pivot_data.sum(axis=0)

# Plotting
fig, ax = plt.subplots(figsize=(12, 8))

# Define bottom for stacking bars
bottom = pd.Series([0] * len(pivot_data), index=pivot_data.index)

# Colors for the bars
colors = plt.cm.tab20.colors

# Plot each store's sales as a stacked bar
for i, store in enumerate(pivot_data.columns):
    bars = ax.bar(pivot_data.index, pivot_data[store], bottom=bottom, color=colors[i % len(colors)], label=f'Store {store} (Total Sales: {store_totals[store]:,.0f})')
    
    # Add text annotations
    for bar in bars:
        height = bar.get_height()
        if height > 0:
            ax.text(
                bar.get_x() + bar.get_width() / 2,
                bar.get_y() + height / 2,
                f'{store}',
                ha='center',
                va='center',
                color='white',
                fontsize=10,
                fontweight='bold'
            )
    
    bottom += pivot_data[store]

# Adding labels and title
plt.xlabel('City')
plt.ylabel('Total Sales (Quantity)')
plt.title('Total Sales by City and Top 5 Stores (Top 5 Cities)')
plt.xticks(rotation=45)
plt.legend(title='Store ID', loc='upper left', bbox_to_anchor=(1, 1))
plt.tight_layout()
plt.show()
```

#### 3.4 Remove some columns that we are not going to use and create new dataframes with partial data.

```python
from datetime import datetime, timedelta

# Select only the specified columns
df_sales_filtered_all = df_sales_filtered[['store_id', 'product_id', 'date', 'sales', 'revenue', 'stock', 'price']]
df_sales_filtered_2018_2019 = df_sales_filtered[df_sales_filtered['date'].dt.year > 2017]
df_sales_filtered_2019 = df_sales_filtered[df_sales_filtered['date'].dt.year > 2018]

# Filter data for the last 6 months
df_sales_filtered_last_6_month = df_sales_filtered[df_sales_filtered['date'] > datetime(2019, 9, 29) - timedelta(days=6*30)]
```

#### 3.5 Group by week and store_id and product_id, year and week

```python
df_sales_filtered_all[(df_sales_filtered_all['store_id'] == 'S0097') & (df_sales_filtered_all['product_id'] == 'P0704')].tail(30)
```

```python
def process_sales_data(df):
     
    # Create new columns for the week of the year and year
    df['week'] = df['date'].dt.isocalendar().week
    df['year'] = df['date'].dt.isocalendar().year
    
    # Drop the date column as it's not needed for grouping
    df = df.drop(columns=['date'])
    
    # Define a custom aggregation function to get the last value
    def last_value(series):
        return series.iloc[-1]
    
    # Group by store_id, product_id, year, and week, then apply aggregations
    df = df.groupby(['store_id', 'product_id', 'year', 'week']).agg({
        'sales': 'sum',
        'revenue': 'sum',
        'stock': last_value,
        'price': last_value  # Use the custom aggregation function for 'price'
    }).reset_index()
    
    return df

df_sales_filtered_all = process_sales_data(df_sales_filtered_all)
df_sales_filtered_2018_2019 = process_sales_data(df_sales_filtered_2018_2019)
df_sales_filtered_2019 = process_sales_data(df_sales_filtered_2019)
df_sales_filtered_last_6_month = process_sales_data(df_sales_filtered_last_6_month)
```
